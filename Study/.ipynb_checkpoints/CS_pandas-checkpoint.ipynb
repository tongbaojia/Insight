```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = []
for y in range(20):
    for i in range(12):
        for c in "abcdefg":
            data.append({"country":c, "GDP":np.random.uniform(0, 1), "time":i, "year": y})
```

```python
df = pd.DataFrame(data)
df_g = df.groupby(['year', 'time'])
```

```python
def fun(a):
    return list(a)[0] + list(a)[-1]
```

```python
df['sum'] = df_g['GDP'].transform(sum)
df['mean'] = df_g['GDP'].transform(np.mean)
df['std'] = df_g['GDP'].transform(np.std)
df['min_GDP'] = df_g['GDP'].transform(min)
df['rank_gdp'] = df_g['GDP'].rank(ascending=False)
df['fun_gdp'] = df_g['GDP'].transform(fun)
df['weighted_gdp'] = df['GDP'] / df['sum']
```
