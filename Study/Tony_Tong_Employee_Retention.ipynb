## Challenge Description
We got employee data from a few companies. We have data about all employees who joined from 2011/01/24 to 2015/12/13. For each employee, we also know if they are still at the company as of 2015/12/13 or they have quit. Beside that, we have general info about the employee, such as avg salary during her tenure, dept, and yrs of experience.

As said above, the goal is to predict employee retention and understand its main drivers

## My Discussions

- The data preparation/feature engineering flows as the following:
    - Find the missing values: if the starting date is two years before the last date of the data, fill the last date of the data as the NA value, then drop NA;
    - Convert depart name into categorical
    - Get a working date as starting date - ending date; save in units of months
    - Split churn into two categories: drop before 17 month and drop after 17 month -- there is a clear separation
    - salary and seniority are normalized;
    - Add in feature month joined; Add in feature salary/seniority;
    - Drop starting date, ending date, depart name,


- Then the problem is treated as a classficiation problem. 
    - The baseline is random;
    - The metric is accuracy; predicting wether an employee will churn in 17 months or not
    - The input features are 6: starting month, department, company, salary, seniority, salary/seniority
    - Methods tried are Logistic regression, Xgboost, Random forest, Neural network, KNN
    - The model is trained with 70% of the data, validated with 15% of data, and tested on the 15% data as reported.
    - Most model behave around 53-58% on the validation set, the best is chosen as the XGboost model, with modest parameter tuning.

```python
from sklearn.model_selection import train_test_split

def pack_data(df, keyword="duration", split=0.15, shuffle=True, SEED = 24):
    '''For training preparation:
    split data, and return a dictionary, 
    containing training, testing and validation
    seed for reproducibility
    '''
    data = {}
    df_X = df.drop(keyword, axis=1).copy()
    df_y = df[keyword].copy()
    X_train, X_test, y_train, y_test = train_test_split(df_X, df_y, train_size=1 - split, test_size=split, random_state=SEED, shuffle=shuffle)
    X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, train_size=1 - split, test_size=split, random_state=SEED, shuffle=shuffle)
    data["X_train"] = X_train.values
    data["y_train"] = y_train.values
    #print(data["y_train"])
    data["X_val"]   = X_val.values
    data["y_val"]   = y_val.values
    data["X_test"]  = X_test.values
    data["y_test"]  = y_test.values
    return data
```

```python
## get data
data = pack_data(df)
y_train  = data["y_train"]
x_train  = data["X_train"] # Creates an array of the train data
x_val    = data["X_val"] # Creats an array of the val data
y_val    = data["y_val"] # Creats an array of the val data
x_test    = data["X_test"] # Creats an array of the test data
y_test    = data["y_test"] # Creats an array of the test data
```

## Setup model train and validation pipeline
### Decide to use accuracy as the metric, with RUC area as a help

```python
def train_valmodel(model):
    model.fit(x_train, y_train)
    ## print("train", metrics.classification_report(y_train, model.predict(x_train)))
    print("train acc %.3f" % metrics.accuracy_score(y_train, model.predict(x_train)))
    print("val acc %.3f" % metrics.accuracy_score(y_val, model.predict(x_val)))
    probas_ = model.predict_proba(x_val)
    # Compute ROC curve and area the curve
    fpr, tpr, thresholds = metrics.roc_curve(y_val, probas_[:, 1])
    plt.plot(fpr, tpr, lw=1, alpha=0.8)
    plt.plot([0, 1], [0, 1], lw=1, alpha=0.8)
    plt.xlabel("fasle positive rate")
    plt.ylabel("true positive rate")
    plt.text(0.7, .25, "ROC area: %.3f" % metrics.auc(fpr, tpr))
    plt.show()
```

```python
lr = LogisticRegression()
train_valmodel(lr)
```

```python
xgbC = xgb.XGBClassifier(max_depth=5, n_estimators=10, learning_rate=0.001, gamma=0.001, subsample=0.6, random_state=0)
train_valmodel(xgbC)
plt.bar(range(len(xgbC.feature_importances_)), xgbC.feature_importances_)
cols = df.drop("duration", axis=1).columns.values
plt.xticks(np.arange(len(cols)), cols)
plt.ylabel("importance of feature")
plt.show()
```

```python
rf = RandomForestClassifier(max_depth=4, random_state=0, n_estimators=200)
train_valmodel(rf)
```

```python
from sklearn.neural_network import MLPClassifier ## NN
nn =  MLPClassifier(hidden_layer_sizes=(100, 20, 5, ), alpha=1)
train_valmodel(nn)
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
train_valmodel(nb)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn =  KNeighborsClassifier(90)
train_valmodel(knn)
```
